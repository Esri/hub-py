```python
from arcgishub.hub import Hub
```

```python
myhub =  Hub("https://www.arcgis.com", "username", "password")
```

We start by fetching the page we would like to edit the layout of.

```python
page1 = myhub.pages.get('e014c56c58854f0289d2c4b4009e1818')
page1
```

```python
layout = page1.layout
```

### Accessing elements of a page

```python
page1.layout.sections[0].ROWS[0].cards[0]
```

### Editing background color of this section



We will now change the background color of this section to a light blue.

```python
layout.sections[1].style.background.color
```

```python
layout.sections[1].style.background.color = '#a0effb'
layout.sections[1].style.background.color
```

### Publish these changes in the layout

```python
page1.update_layout(layout)
```

### Adding a new card to the page

To add a new `card` you update the `cards` list for the row within the section with the dictionary for the new card.

We start by first accessing the row of interest we would like to add the new card to.


```python
layout.sections[0].rows[0]
```

The first row of the first section has 1 card. We will now add another one.

```python
new_card = {'component': {'name': 'markdown-card', 'settings': {'markdown': '<h2 class="text-center"><strong><br></strong></h2><h2 class="text-center" style="text-align: left; "><strong>About the Data</strong></h2>\n<h4 style="text-align: left;">Use this section to provide additional context for your dataset.</h4><h4 style="text-align: left;"><br> For instance, useful details that can be provided to users are background and limitations or PSA of the dataset, attributes and fields provided, details on how it is collected and how often it is updated, link to metadata, etc</h4><p class="text-center" style="text-align: left;"><br></p><p class="text-center"><br></p>', 'schemaVersion': 1}}, 'width': 9, 'showEditor': False}

layout.sections[0].rows[0].cards.append(new_card)
layout.sections[0].rows[0].cards
```

We will now publish these changes and verify the new card added

```python
page1.update_layout(layout)
```

As we notice below, a new card successfully got added in the same row



### Deleting a card from the page

We will now delete the same card from the page

```python
layout.sections[0].rows[0].cards.pop(-1)
```

```python
page1.update_layout(layout)
```

As we see, the card has been successfully deleted from the page.
