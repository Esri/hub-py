```python
from arcgishub.hub import Hub
myhub = Hub("https://www.arcgis.com", "mmajumdar_dcdev")
```

```python
from arcgis.gis import GIS
gis = GIS("https://www.arcgis.com", "mmajumdar_dcdev")
```

### We start by creating a new initiative

```python
i1 = myhub.initiatives.add('New initiative1')
i1.item
```

### We now add content to the initiative content library

```python
content = gis.content.get('13261595742e49e78bec358667f86e4b')
content
```

```python
i1.add_content(content)
```

### We now reassign this initiative to another user, i.e. we change ownership of the initiative and all its content

```python
i1.reassign_to('manushi.majumdar')
```

#### We notice that the initiative has been transferred to the new owner. We also verify that the contents (webmap) has been transferred to the new owner below

```python
gis.content.get('13261595742e49e78bec358667f86e4b')
```
